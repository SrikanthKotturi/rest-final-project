```python

import pandas as pd
from sqlalchemy import create_engine

# Database connection URL
DATABASE_URL = "postgresql://postgres:postgres@localhost:5432/postgres"

# Create an engine instance
engine = create_engine(DATABASE_URL)

# Read data from PostgreSQL into a pandas DataFrame
query = "SELECT * FROM patients"
df = pd.read_sql(query, engine)

# Display the DataFrame
df.head()
```

```python
import matplotlib.pyplot as plt

# Plot configurations
plt.figure(figsize=(15, 10))

# Age distribution
plt.subplot(2, 2, 1)
df['age'].plot(kind='hist', bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Gender distribution
plt.subplot(2, 2, 2)
df['gender'].value_counts().plot(kind='bar', color=['lightcoral', 'lightskyblue'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')

# Blood type distribution
plt.subplot(2, 2, 3)
df['blood_type'].value_counts().plot(kind='bar', color='mediumseagreen')
plt.title('Blood Type Distribution')
plt.xlabel('Blood Type')
plt.ylabel('Count')

# Admission type distribution
plt.subplot(2, 2, 4)
df['admission_type'].value_counts().plot(kind='bar', color='mediumpurple')
plt.title('Admission Type Distribution')
plt.xlabel('Admission Type')
plt.ylabel('Count')

plt.tight_layout()
plt.show()

# Additional visualizations for billing amount and medical conditions
plt.figure(figsize=(15, 5))

# Billing amount distribution
plt.subplot(1, 2, 1)
df['billing_amount'].plot(kind='hist', bins=30, color='peachpuff', edgecolor='black')
plt.title('Billing Amount Distribution')
plt.xlabel('Billing Amount')
plt.ylabel('Frequency')

# Medical condition distribution
plt.subplot(1, 2, 2)
df['medical_condition'].value_counts().plot(kind='bar', color='lightsteelblue')
plt.title('Medical Condition Distribution')
plt.xlabel('Medical Condition')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='medical_condition', y='billing_amount', palette='pastel')
plt.title('Billing Amount per Medical Condition')
plt.xlabel('Medical Condition')
plt.ylabel('Billing Amount')
plt.xticks(rotation=45)
plt.show()
```
